```python
def menu():
    print('1.Create New User.')
    print('2.Show User.')
    print('3.Delete User.')
    print('4.Deposit money.')
    print('5.Withdraw Money.')
    print('6.Transfer MOney.')
    print('7.Show All Users.')
    print('8.Exit.')


def getting_info():
    first_name = input('Enter Your First Name: \n')
    last_name = input('Enter Your Last Name: \n')
    id = id_generator()
    password = password_verify()
    amount = int(input('Enter Your Initial Amount: \n'))
    user = customers(first_name = first_name, last_name = last_name, id = id, password = password, amount = amount)
    session.add(user)
    session.commit()
    print('User Created Successfully!\n')

    
def id_generator():
    id = random.randint(999,10000)
    users = session.query(customers)
    for user in users:
        if (user.id) != id:
            return id
        else:
            print('ID Already Exists!\n')                 
        
    
def transfer():
    confirm_flag = False
    insufficient_flag = True
    giver = user_verify()
    reciever = id_verify()
    while insufficient_flag:
        amount = int(input('Enter The Amount You Want To Transfer: \n'))
        if giver[4] >= amount:
            updated_amount_giver = giver[4] - amount
            giver_data = session.query(customers).filter(customers.id == giver[0]).first()
            giver_data.amount = updated_amount_giver
            session.commit()
            updated_amount_reciever = reciever[4] + amount
            recievers_data = session.query(customers).filter(customers.id == reciever[0]).first()
            recievers_data.amount = updated_amount_reciever
            session.commit()
            confirm_transfer = True
            insufficient_flag = False
            break
        else:
            print('Invalid Amount Entered!\n')
            insufficient_flag = True
    if confirm_transfer:
        print('Transfer Successful!\n')


def withdraw():
    invalid_amount_flag = True
    confirm_withdraw = False
    data = user_verify()
    while invalid_amount_flag:
        amount = int(input('Enter The Amount You Want To Deposit: \n'))
        if data[4] >= amount:
            updated_amount = data[4] - amount
            users = session.query(customers).filter(customers.id == data[0]).first()
            users.amount = updated_amount
            session.commit()
            invalid_amount_flag = False
            confirm_withdraw = True
        else:
            print('Invalid Amount Entered!\n')
            invalid_amount_flag = True
    if confirm_withdraw:
        print('Amount Successfully Withdrawed!\n')

def deposit():
    data = user_verify()
    amount = int(input('Enter The Amount You Want To Deposit: \n'))
    updated_amount = data[4] + amount
    users = session.query(customers).filter(customers.id == data[0]).first()
    users.amount = updated_amount
    session.commit()
    print('Amount Deposited Successfully!\n')
    

def id_verify():
    id_confirm = False
    while True:
        id = int(input('Enter recievers id: \n'))
        users = session.query(customers)
        for user in users:
            if user.id == id:
                print('User Found!\n')
                return user.id,user.first_name,user.last_name,user.password,user.amount
        else:
            print('Invalid ID!\n')
            id_confirm = True


def user_verify():
    user_found = False
    while True:
        id = int(input('Enter Your ID: \n'))
        password = input('Enter Your Password: \n')
        users = session.query(customers)
        for user in users:
            if user.id == id and user.password == password:
                print('User Found!\n')
                return user.id,user.first_name,user.last_name,user.password,user.amount
        else:
            print('Invalid Credentials!\n')
            user_found = True


def show_user():
    user = user_verify()
    print(user)
    
    
def delete_user():
    user = user_verify()
    deleted_item = session.query(customers).filter(customers.id == user[0]).first()
    session.delete(deleted_item)
    session.commit()
    print('User Successfully Deleted!\n')


def show_all():
    users = session.query(customers)
    for user in users:
        print(f"id = {user.id},first_name = {user.first_name},last_name = {user.last_name},password = {user.password},amount = {user.amount}")

            
def password_verify():
    pattern = re.compile(r'^(?=.*[A-Z])(?=.*[a-z])(?=.*[!@#$%^&*])(?=.*\d).+$')
    while True:
        password = input('Enter Strong Password: \n')
        if len(password) >= 8:
            if pattern.match(password):
                return password
            else:
                print("\nPassword should contain at least:\n1 special character\n1 number!\n1 uppercase letter\n"
                          "1 lowercase letter\n")
        else:
            print("\nPassword length less then 8 characters\n")
```
